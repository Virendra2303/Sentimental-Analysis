**Importing Data**

```python
train_df=pd.read_csv('training.txt',sep='	')
```

**Chaning Column Name**

```python
cols_df=['Rating','Review']
train_df.columns=cols_df
```

Word Embedding Require 2 steps:
1.   **Tokenizing**(converting the words to some number)
2.   **Padding**(to have Fix size of input)



```python
vocab_len=1000
def tokenizing(df,colmn):
  token_df=[tf.keras.preprocessing.text.one_hot(i,vocab_len) for i in df[colmn]]
  return token_df
```

```python
token_df=tokenizing(train_df,'Review')
token_df[:5]
```

To Find out the maximum length of feature vector

```python
max_length=0
for i in token_df:
  if len(i)>max_length:
    max_length=len(i)
max_length
```

```python
def padding(df,max_length=max_length):
 padded_df=tf.keras.preprocessing.sequence.pad_sequences(df,maxlen=max_length,padding='post')
 return padded_df
```

```python
padded_df=padding(token_df)
padded_df
```

Spiltting Dataframe into Train and Test

```python
train=padded_df[0:int(0.8*(len(padded_df)))]
test=padded_df[int(0.8*(len(padded_df))):]
```

**Training Model**

```python
embedding_len=5
model=tf.keras.models.Sequential([
    tf.keras.layers.Embedding(vocab_len,embedding_len,input_length=max_length,name='embedding'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(36,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(train,train_df['Rating'].head(int(0.8*(len(padded_df)))),epochs=20)
```

Evaluating score for test data

```python
model.evaluate(test,train_df['Rating'].iloc[int(0.8*(len(padded_df))):])
```

**Testing Model for the Unseen data**

```python
test_df = pd.read_csv('/content/testdata.txt', sep='     ', engine='python',names=['Review'])
test_df
```

```python
test_token_df=tokenizing(test_df,'Review')
test_padded_df=padding(test_token_df)
```

```python
y_predict_sigmoid_test=model.predict(test_padded_df)
```
